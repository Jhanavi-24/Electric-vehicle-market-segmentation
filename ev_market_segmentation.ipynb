###Conclusion
Based on their metrics, the top companies for electric car manufacturing are Audi, followed by Mercedes and then Mahindra. However, it can be inferred that not all criteria have the highest scores for each brand. For instance, while the Audi e-tron GT has the best range compared to other brands, its charging capacity is not as impressive. To gain a competitive advantage, it would be beneficial to focus on improving the median performance of two key criteria, namely range and charging, surpassing that of Audi. By identifying and addressing such gaps in existing brands, it becomes possible to rise above the competition and potentially outperform them.

```python
customers_behaviour = pd.read_csv("/content/Indian automoble buying behavour study 1.0.csv")
```

```python
customers_behaviour.head()
```

```python
customers_behaviour.info()
```

```python
plt.figure(1,figsize=(15,7))
n = 0
for cols in ['Age','No of Dependents','Salary','Wife Salary','Total Salary','Price']:
  n += 1
  plt.subplot(1,6,n)
  sns.set(style = 'whitegrid')
  plt.subplots_adjust(hspace=0.5,wspace=0.5)
  sns.swarmplot(x= cols, y = 'Make', data=customers_behaviour)
  plt.ylabel("Make" if n==1 else '')
  plt.title('Swarm Plot')
plt.savefig("TEst.jpg")
```

###Conclusion
Younger individuals tend to opt for lower-priced cars, but as their income increases with age, they are more likely to choose more expensive vehicles. Moreover, it can be observed that as the number of dependents increases, there is a preference for cars with spacious interiors. This is evident by the higher frequency of SUV purchases among individuals with more dependents. Additionally, when the wife is employed and earns a substantial income, people tend to buy pricier cars, as the household's overall income is higher compared to households where the wife is a homemaker.

```python
ev_stas = ev_stats.set_index("State")
columns = ev_stats.columns[2:-1]
columns
```

```python
count = 2
for z in columns:
  sns.barplot(x="State",y=z,data=ev_stats,order=ev_stats.sort_values(z).State)
  sns.color_palette("pastel")
  plt.xticks(rotation=90)
  plt.ylabel('')
  plt.title(f"{z}, hightest = {plt.xticks()[1][-1]}")
  plt.savefig(f"output-{count}.jpg",bbox_inches='tight')
  count+=1
  plt.show()
```

```python
sns.barplot(x="State",y='Total in state',data=ev_stats,order=ev_stats.sort_values("Total in state").State)
sns.color_palette("husl")
plt.xticks(rotation=90)
plt.title(f"Total ev vehicles, hightest = {plt.xticks()[1][-1]}")
plt.savefig("total.jpg",bbox_inches='tight')
plt.show()
```

###Conclusion
Based on the information provided, we can deduce that Maharashtra, Uttar Pradesh, Gujarat, and Delhi are the four states with the highest adoption of electric vehicles. This indicates a strong level of competition among these states and a growing interest among their respective populations in electric vehicles.
